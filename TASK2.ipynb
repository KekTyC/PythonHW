# Задание 2.14 (8 баллов)

Постройте график функции $$f(x) = (5  \sin(2x) - \cos^2(x) - 1 + |x^3| - x^2) / (x^2 + 1)$$  

и производной от этой функции.  Найдите минимум с помощью `scipy` и обозначьте его на графике. На этой же оси постройте графики скользящего среднего этой функции с окном $1$, $2$, $5$.

```python
from scipy.misc import derivative
from scipy.optimize import minimize
plt.figure(figsize=(17, 9))

def f(x):
    return (5 * np.sin(2 * x) - np.cos(x) ** 2 - 1 + np.abs(x ** 3) - x ** 2) / (x ** 2 + 1)

x = np.linspace(-10, 10, 100)
minimum = minimize(f, 0)

plt.plot(x, f(x), label="Function")
plt.plot(x, derivative(f, x), label="Derivative")
plt.scatter(minimum.x, minimum.fun, label="Minimum", color="black")
plt.plot(x, movav(f(x), 1), label="Moving Average 1")
plt.plot(x[1:], movav(f(x), 2), label="Moving Average 2")
plt.plot(x[4:], movav(f(x), 5), label="Moving Average 5")
plt.legend()
plt.show()   
```

# Задание 2.15 (10 баллов)

Есть следующий алгоритм семплинирования, на вход функции подается вектор из целых положительных чисел и число семплов.

```python
from typing import List 

def sample(x: List[int], c: int) -> List[int]:
    assert len(x) > 0
    
    s = np.sum(x)
    res = []
    for _ in range(c):
        val = s * np.random.random()
        cur, idx = 0, 0        
        while cur + x[idx] <= val:
            cur += x[idx]
            idx += 1
            
        res.append(idx)
    return res

            
sample([50, 3, 1, 7, 20], 5)   
```

Реализуйте данный алгоритм более оптимальным способом на массивах `numpy`, без использования циклов. Проверьте корректность и идентичность алгоритмов семплирования с помощью статистических тестов.

```python
def upgraded_sample(x, c):
    return np.random.choice(len(x), c, p = np.array(x) / np.array(x).sum())
```

```python
from scipy.stats import ttest_ind
test = np.random.randint(0, 100, size=(1000, ))
ttest_ind(sample(test, 1000), upgraded_sample(test, 1000))
# Нет основания отвергать нулевую гипотезу о равенстве средних
```

# Задание 2.16 (5 баллов)

Есть следующий алгоритм "случайного блуждания"

```python
import random 

def walk(n: int):
    cur_position = 0
    for i in range(n):
        cur_position += random.choice([-3, -1, 0, 2, 3])

    return cur_position

walk(10_000)
```

Реализуйте более эффективную версию с помощью `numpy`, сравните производительность (используя, например, magic `timeit`). Воспользовавшись [kdeplot()](https://seaborn.pydata.org/generated/seaborn.kdeplot.html) или аналогами, постройте график распределения значений `walk(10_000)` на каком-то разумном числе запусков. 

```python
def upgraded_walk(n):
    return np.random.choice([-3, -1, 0, 2, 3], size=n).sum()
```

```python
%%timeit
upgraded_walk(10_000)
```

```python
%%timeit
walk(10_000)
```

```python
import seaborn as sns
sns.set(rc={'figure.figsize':(17,10)})
sns.kdeplot([upgraded_walk(10_000) for i in range(500)])
sns.kdeplot([walk(10_000) for i in range(500)])
plt.show()
```

# Задание 2.17 (5 баллов)

Имеется функция, которая проделывает достаточно странные манипуляции с массивом `numpy`

```python
import numpy.typing as npt

def strange_conv(a: npt.NDArray[np.float64]) -> np.float64:
    x = a.shape[0]
    y = a.shape[1]

    t = 1.
    for y in range(y - 5):
        for x in range(x - 3):
            c = 1.5 * a[y+1, x+2] - a[y+5, x+3] * a[y, x] + 0.2 * a[y+4, x]
            t = 0.2 * t + 0.8 * c

    return t


a = np.random.random((10000, 10000))
strange_conv(a)
```
